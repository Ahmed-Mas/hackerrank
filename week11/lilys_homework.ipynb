```python
# this works in O(n logn) we get a sorted array, and make swaps using indexes to match the sorted array
# then just count the swaps as we go through it.
# on every swap, the ith index will be correct, and we just go through the array once O(n)
# but cause we had to pre-sort the array earlier O(log n), its O(n log n)

def count_swaps(arr, sor):
    arr = arr.copy()
    idx_dict = {val: idx for idx, val in enumerate(arr)}

    swaps = 0
    for idx in range(len(arr)):
        if arr[idx] != sor[idx]:
            swaps += 1
            correct_idx = idx_dict[sor[idx]]
            idx_dict[arr[idx]] = correct_idx
            arr[idx], arr[correct_idx] = sor[idx], arr[idx]
    return swaps
```
